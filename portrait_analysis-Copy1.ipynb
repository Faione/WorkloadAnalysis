```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = {
    'group1': {
        'col1': [25, 30, 35, 40, 45, 50, 55, 60, 65],
        'col2': [35, 38, 42, 48, 55, 60, 68, 72, 78]
    },
    'group2': {
        'col1': [20, 25, 30, 35, 40, 45, 50, 55, 60],
        'col2': [30, 32, 35, 38, 40, 42, 45, 50, 55]
    },
    'group3': {
        'col1': [15, 20, 25, 30, 35, 40, 45, 50, 55],
        'col2': [25, 28, 30, 32, 35, 38, 40, 42, 45]
    }
}

# 数据转换
df_list = []
for group, cols in data.items():
    for col, values in cols.items():
        temp_df = pd.DataFrame({
            'Group': [group] * len(values),
            'Column': [col] * len(values),
            'Value': values
        })
        df_list.append(temp_df)
df = pd.concat(df_list, ignore_index=True)

# # 使用 seaborn 画图
# plt.figure(figsize=(12, 6))

# for i in range(0,3):
#     sns.boxplot(x="Group", y="Value", hue="Column", data=df, dodge=True)

# plt.xticks(rotation=45)
# plt.tight_layout()
# plt.show()


pic_ct = 3
figsize=(10,pic_ct*2)

# 1. 创建子图
fig, axs = plt.subplots(pic_ct, sharex=True, figsize=figsize)

# 2. 在每个子图上绘制指标
for i in range(0,pic_ct):
    sns.boxplot(x="Group", y="Value", hue="Column", data=df.copy(), dodge=True)

    axs[i].set_xlabel("x-label")
    axs[i].set_ylabel("y-label")

# 3. 显示图形
plt.title("kk")
plt.tight_layout()
# plt.legend(loc='upper left')
plt.show()
```
