```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = [
    [25, 30, 35, 40, 45, 50, 55, 60, 100],
    [35, 38, 42, 48, 55, 60, 68, 72, 78]
]

figure,axes=plt.subplots() #得到画板、轴
axes.boxplot(data,
             patch_artist=True, # 箱体上色
             widths = 0.1, # 箱宽
             notch = True, # 凹口
             showmeans = True, # 均值
             positions=[1,4], # 箱体位置
             
            ) 
plt.show()
```
